```python
# A. Write a program
# Step 1: Create a  condition = 'LoanStatus'== 'Completed'| 'LoanStatus' == 'Defaulted' |\'LoanStatus' == 'Chargedoff'

# Step2: Create a user define function  using condition and `LoanStatus` column.

# Hint: df['LoanStatus'] = df.apply(user define function , axis=1)

```

```python
df = df[df.apply(lambda x : x['LoanStatus']== 'Completed' or x['LoanStatus'] == 'Defaulted' or
            x['LoanStatus'] == 'Chargedoff', axis=1)]
```

```python
df['LoanStatus']  = df.apply(lambda x: 'Defaulted' if x['LoanStatus']=='Chargedoff' else x['LoanStatus'],axis=1)
```

```python
df['LoanStatus'].value_counts()
```

```python
# B. Write a program
# Step 1: Create a  dictionary called categories = 1: 'Debt Consolidation', 2: 'Home Improvement', 
# 3: 'Business', 6: 'Auto', 7: 'Other'

# Step2: Create a user define function  using categories and `ListingCategory (numeric)` column.

# Hint: df['ListingCategory (numeric)'] = df.apply(user define function , axis=1)
```

```python
cat  = {1: 'Debt Consolidation', 2: 'Home Improvement',  3: 'Business', 6: 'Auto', 7: 'Other'}

def rep(x):
    if x>7:
        return 'Other'
    else:
        return pd.Series(x).replace(cat).values[0]
```

```python
df['ListingCategory (numeric)'] = df.apply(lambda x : rep(x['ListingCategory (numeric)']),axis=1)
```

```python
df['ListingCategory (numeric)'].value_counts()
```

```python
# C. Display the box plot for ProsperRating (Alpha) vs 
# LoanOriginalAmount and hue = Loan Status[Completed,Defaulted]. Write your observations.
```

```python
plt.figure(figsize=(20,7))
sns.boxplot(data=df,y='LoanOriginalAmount',x='ProsperRating (Alpha)',hue='LoanStatus')

# Defaulted has very negligible IQR in HR,A,B prosper rating categories
# Defaulted loan original amount is having a smalller range compared to completed in the AA category
```

```python
# D. Display the catplot for ProsperRating (Alpha) vs ListingCategory (numeric)[Debt Consolidation,Other,Business,
#   Home Improvement,Auto] and hue = Loan Status[Completed,Defaulted]. Write your observations.
```

```python
plt.figure(figsize=(15,7))
sns.catplot(data=df,x='ProsperRating (Alpha)',y='ListingCategory (numeric)',hue='LoanStatus',kind='swarm',height=7, aspect=1.5)

# Defaulted are mostly identified in Hr, D and E prosper rating
# Defaulted are mostly identified in Other and Debt Consolidation listing category
```
